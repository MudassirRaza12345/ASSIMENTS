```python
#overloading :compile time run error hota ha
#isma same class do same  method bngae 
#same naam ka method ho pr is ma parameter different ho
#yani parameter jo hum pas kr rha ho wo ziada ho()
#for suppose we have sum(self,a,b) and (self,a,b,c)
class overloadingDemo():
    def sum(self,a=None,b=None,c=None):
        if a!= None and b!= None and c!=None:
            print(a+b+c)
        elif a!= None and b!= None:
            print(a+b)
        
obj3 = overloadingDemo()

obj3.sum(2,4,8)
obj3.sum(2,4)
```

```python
#e.gPolymorphism
class Karachi():
    def detail(self):
        print("The Largest city of Pakistan is Karachi")
class Lahore():
    def detail(self):
        print("The Heart of city of Pakistan is Lahore")
class Islamabad():
    def detail(self):
        print("The capital of city of Pakistan is Islamabad")
class Sialkot():
    def detail(self):
        print("The sport  city of Pakistan is Sialkot")
    
def About_Cities(Cities):
    Cities.detail()
    
k = Karachi()
L = Lahore()
I = Islamabad()
S = Sialkot()

About_Cities(k)
print("\n")
About_Cities(L)
print("\n")
About_Cities(I)
print("\n")
About_Cities(S)
print("\n")
```

```python
#Abstraction
#humko kisi bhi cheez ki comlexity ka bara maaloom na ho ka wo kis tarh 
#work krhi ha suppose kuch function bna ho hum is functionality ko 
#used kr rha ha baqi iska bara humko kuch nhi maloom ho
#for suppose hamara ps television ha ab kis trh sa backhan pr work kr rha 
#hum ko nhi maaloom bs hum samna hi remote sa change krha hota ha
#mobile we used but cant about its backhand work
#is tarh abstraction ma bhi asa hota ha ka aik hum aik abstract class banata ha 
#pr is ma hum complete detail nhi bataa.isam hum method ka naam batadeta ha 
#Abstract class(main class) ko hum clean rekhta ha sirh function or
#method batata ha.
#abstract ka lya  zaroori abstract based class(abc) ko import kra or
#abstract method ko import kra jo anaconda ka sath install ho jati ha
#abc hamra ps aik module ha
#Abstraction ma  lazmi ha child class parent class ko call kra
# ya used iska bagair error 
#child class ka zaroori ha ka parent class ko used kra
#is ko decorater/decoration kehta ha @abstractmethod


from abc import ABC,abstractmethod
#Nikkah is Abstract class
class Nikkah(ABC):
    @abstractmethod
    def haqmahar(self):
        pass
class human1(Nikkah):
    def haqmahar(self,tola):
        print("he gives "+str(tola)+" tola")
      
    def valima(self):
        print("valima going on")
    
    
    
    
obj6=human1()
obj6.haqmahar(4)
obj6.valima()
    
```
